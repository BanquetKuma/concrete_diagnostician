```python
# 所定のフォルダ以下のdocをdocxに変換

import os
from doc2docx import convert

def convert_doc_to_docx(folder_path):
    converted_files = []
    errors = []

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith('.doc'):
                doc_path = os.path.join(root, file)
                docx_path = os.path.splitext(doc_path)[0] + '.docx'
                
                try:
                    convert(doc_path, docx_path)
                    converted_files.append(docx_path)
                    print(f"変換成功: {doc_path} -> {docx_path}")
                except Exception as e:
                    errors.append((doc_path, str(e)))
                    print(f"変換失敗: {doc_path} - エラー: {str(e)}")

    return converted_files, errors

# 変換を実行
converted, errors = convert_doc_to_docx(doc_folder_path)

print(f"\n変換されたファイル数: {len(converted)}")
print(f"エラーが発生したファイル数: {len(errors)}")

if errors:
    print("\nエラーが発生したファイル:")
    for doc_path, error_msg in errors:
        print(f"{doc_path}: {error_msg}")
```

```python
# 所定のフォルダ以下のdocxをpdfに変換

import os
import logging
from docx2pdf import convert

def convert_docx_to_pdf(folder_path):
    converted_files = []
    errors = []

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith('.docx'):
                docx_path = os.path.join(root, file)
                pdf_path = os.path.splitext(docx_path)[0] + '.pdf'
                
                try:
                    logging.info(f"WordファイルをPDFに変換中: {docx_path}")
                    convert(docx_path, pdf_path)
                    converted_files.append(pdf_path)
                    logging.info(f"変換成功: {docx_path} -> {pdf_path}")
                except Exception as e:
                    errors.append((docx_path, str(e)))
                    logging.error(f"{docx_path}のPDF変換中にエラーが発生しました: {str(e)}")

    return converted_files, errors

# ログの設定
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')


# 変換を実行
converted, errors = convert_docx_to_pdf(doc_folder_path)

print(f"\n変換されたファイル数: {len(converted)}")
print(f"エラーが発生したファイル数: {len(errors)}")

if errors:
    print("\nエラーが発生したファイル:")
    for docx_path, error_msg in errors:
        print(f"{docx_path}: {error_msg}")
```

```python
# Blob StorageにPDFをアップロード（フォルダ構造を維持）

import os
from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient

def upload_pdfs_to_blob(local_folder_path, connection_string, container_name):
    # Blobサービスクライアントを作成
    blob_service_client = BlobServiceClient.from_connection_string(connection_string)

    # コンテナクライアントを取得
    container_client = blob_service_client.get_container_client(container_name)

    # ローカルフォルダ内のPDFファイルを再帰的に検索してアップロード
    for root, dirs, files in os.walk(local_folder_path):
        for file in files:
            if file.lower().endswith('.pdf'):
                local_file_path = os.path.join(root, file)
                
                # Blob内のパスを作成（フォルダ構造を維持）
                relative_path = os.path.relpath(local_file_path, local_folder_path)
                blob_path = relative_path.replace("\\", "/")  # Windowsパスを修正

                # BlobClientを取得
                blob_client = container_client.get_blob_client(blob_path)

                # ファイルをアップロード
                with open(local_file_path, "rb") as data:
                    blob_client.upload_blob(data, overwrite=True)
                
                print(f"アップロード完了: {local_file_path} -> {blob_path}")

# 実行
connection_string = "DefaultEndpointsProtocol=https;AccountName=kmakicistrage2;AccountKey=a460yXb+uyR0pZNC9ugy5LiLVKV4WvW7mqYxWjikLydKkm2OSJwOn7rR0mDmVKin/IEyd/hLWcir+AStlKazsg==;EndpointSuffix=core.windows.net"
container_name = "20240928-rag-test-container"
# upload_pdfs_to_blob(doc_folder_path, connection_string, container_name) # 本番実行時はコメントアウトを外す
```

```python
# Blob Storageに保存されたPDFファイルのOCRを実行

import os
import tempfile
from azure.storage.blob import BlobServiceClient
import pypdf

def perform_ocr(blob_url, split_f_path, extract_f_path, chunk_size, extract_keywords, new_file_name):
    """doc string
    blob_url: str OCRを実行するBlobのURL
    split_f_path: str PDFを分割する際の出力ディレクトリのパス
    extract_f_path: str 抽出したテキストを保存する際の出力ディレクトリのパス
    chunk_size: int ページ数ごとにPDFを分割する際のページ数
    extract_keywords: list 抽出するキーワードのリスト
    new_file_name: str 出力ファイル名の接頭辞"""
    # BlobServiceClientの作成
    blob_service_client = BlobServiceClient(account_url="https://pdfalldata.blob.core.windows.net")
    
    # コンテナとBlobの名前を取得
    container_name = blob_url.split('/')[3]
    blob_name = '/'.join(blob_url.split('/')[4:])
    print(f"コンテナ名: {container_name}, Blob名: {blob_name}")
    
    # BlobClientの取得
    blob_client = blob_service_client.get_container_client(container_name).get_blob_client(blob_name)
    
    # 一時ファイルの作成
    with tempfile.NamedTemporaryFile(delete=False, suffix='.pdf') as temp_file:
        temp_file_path = temp_file.name
        
        # Blobのダウンロードと一時ファイルへの書き込み
        blob_data = blob_client.download_blob()
        blob_data.readinto(temp_file)
        print(f"一時ファイル {temp_file_path} を作成しました。")
    
    try:
        # PDFの読み込み（確認用）
        pdf_reader = pypdf.PdfReader(temp_file_path)
        total_pages = len(pdf_reader.pages)
        print(f"Total pages in PDF: {total_pages}")
        
        # ここでOCR処理を行う（DocumentIntelligenceClientを使用）
        client = DocumentIntelligenceClient(endpoint="https://rag-test-di.cognitiveservices.azure.com/",
                                        key="5Bu5fhh7BzoLNl0HjWXmYOeDTCqQ60EJr5fLaMN0LUukhIjoyd7UJQQJ99ALACi0881XJ3w3AAALACOG9O9g",
                                        api_version="2024-07-31-preview")
        client.custom_flag = False # カスタムモデルを使用しない場合はFalse
        result = client.get_processed_result(temp_file_path, split_f_path, extract_f_path, chunk_size, extract_keywords, new_file_name)
        
        return result
    
    finally:
        # 処理が終了したら（エラーが発生しても）一時ファイルを削除
        if os.path.exists(temp_file_path):
            os.remove(temp_file_path)
            print(f"一時ファイル {temp_file_path} を削除しました。")
```

```python
import pandas as pd

# OCR処理結果を分類してDataFrameに追加する関数
def classify_and_append_to_df(result, df):
    table_data = []
    figure_data = []
    other_texts = []
    title_data = []
    
    for entry in result:
        content_type = entry['type']
        if content_type == '図':
            figure_data.append(str(entry))  # コンテンツの代わりにJSON全体を追加
        elif content_type == '表':
            table_data.append(str(entry))  # コンテンツの代わりにJSON全体を追加
        elif content_type == 'その他の文章':
            other_texts.append(str(entry))  # コンテンツの代わりにJSON全体を追加
        else:
            title_data.append(str(entry))# コンテンツの代わりにJSON全体を追加
    # 内容が空の場合にnote_dataを追加するロジックは削除。必要に応じてここに追加可能。
    
    # DataFrameに追加
    df = pd.concat([df, pd.DataFrame([{
        "filename": result[0]['file_name'] if result else '',
        #"majorcategory": daibunrui,
        #"subcategory": chubunrui,
        "title": title_data,
        "content": other_texts,
        "tableinfo": table_data,
        "graphinfo": figure_data
    }])], ignore_index=True)

    return df
```

```python
"""
This code sample shows Prebuilt Document operations with the Azure Form Recognizer client library. 
The async versions of the samples require Python 3.6 or later.

To learn more, please visit the documentation - Quickstart: Form Recognizer Python client library SDKs
https://learn.microsoft.com/azure/applied-ai-services/form-recognizer/quickstarts/get-started-v3-sdk-rest-api?view=doc-intel-3.1.0&pivots=programming-language-python
"""

from azure.core.credentials import AzureKeyCredential
from azure.ai.formrecognizer import DocumentAnalysisClient

"""
Remember to remove the key from your code when you're done, and never post it publicly. For production, use
secure methods to store and access your credentials. For more information, see 
https://docs.microsoft.com/en-us/azure/cognitive-services/cognitive-services-security?tabs=command-line%2Ccsharp#environment-variables-and-application-configuration
"""

endpoint = "https://test20240928-kmakici.cognitiveservices.azure.com/"
key = "6b4bc8ebcb10472aa517a7f04c7fa594"

# sample document
formUrl = "https://kmakicistrage2.blob.core.windows.net/20240928-rag-test-container/CONVENTION/オーストラリアシドニー.pdf"

document_analysis_client = DocumentAnalysisClient(
        endpoint=endpoint, credential=AzureKeyCredential(key)
    )
    
poller = document_analysis_client.begin_analyze_document_from_url("prebuilt-document", formUrl)
result = poller.result()

print("----Key-value pairs found in document----")
for kv_pair in result.key_value_pairs:
    if kv_pair.key and kv_pair.value:
        print("Key '{}': Value: '{}'".format(kv_pair.key.content, kv_pair.value.content))
    else:
        print("Key '{}': Value:".format(kv_pair.key.content))

print("----------------------------------------")
```
