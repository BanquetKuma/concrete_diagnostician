```python
from sudachipy import Dictionary, SplitMode, Tokenizer
from sudachipy.morpheme import Morpheme
from typing import Literal
import ast

class SudachiNounNormalizer:
    """google style docstring
    A class to normalize Japanese nouns using SudachiPy."""
    SPLIT_MODES = {
        "A": SplitMode.A,
        "B": SplitMode.B,
        "C": SplitMode.C
    }
    MAX_LENGTH = 18000

    def __init__(
            self,
            dict_type: Literal["small", "core", "full"] = "core",
            split_mode: Literal["A", "B", "C"] = "C")-> None:
        """google style docstring
        Args:
            dict_type (Literal["small", "core", "full"], optional): The type of dictionary to use. Defaults to "core".
            split_mode (Literal["A", "B", "C"], optional): The split mode to use. Defaults to "C".
        """
        self._dictionary = Dictionary(dict=dict_type)
        self._tokenizer = self._dictionary.create(self.SPLIT_MODES[split_mode])
        self._morphemes = self.tokenizer.tokenize('')
        self.is_noun = self._dictionary.pos_matcher([("名詞",)])

    @property
    def dictionary(self) -> Dictionary:
        """google style docstring
        Returns:
            Dictionary: The dictionary object."""
        return self._dictionary

    @property
    def tokenizer(self) -> Tokenizer:
        """google style docstring
        Returns:
            Tokenizer: The tokenizer object."""
        return self._tokenizer
    # 名詞のみを正規化する
    def morpheme2norm(self, morpheme: Morpheme) -> str:
        return morpheme.normalized_form() if self.is_noun(morpheme) else morpheme.surface()

    def normalize(self, document: str) -> str:
        """google style docstring
        Normalizes a document.
        Args:
            document (str): The document to normalize.
        Returns:
            str: The normalized document."""
        if not isinstance(document, str):
            raise ValueError("Input must be a string.")
        sub_documents = (document[i:i+self.MAX_LENGTH] for i in range(0, len(document), self.MAX_LENGTH))
        return "".join(
            (
                "".join((
                    self.morpheme2norm(morpheme)
                    for morpheme in self.tokenizer.tokenize(sub_document, out=self._morphemes)
                ))
                for sub_document in sub_documents
            )
        )

    def extract_corrected_words(self, document: str) -> list:
        """google style docstring
        Extracts corrected words from a document.
        Args:
            document (str): The document to extract corrected words from.
        Returns:
            list: A list of tuples containing the original word and the corrected word."""
        if not isinstance(document, str):
            raise ValueError("Input must be a string.")
        sub_documents = (document[i:i+self.MAX_LENGTH] for i in range(0, len(document), self.MAX_LENGTH))
        corrected_words = []
        for sub_document in sub_documents:
            for morpheme in self.tokenizer.tokenize(sub_document, out=self._morphemes):
                normalized_form = self.morpheme2norm(morpheme)
                if self.is_noun(morpheme) and normalized_form != morpheme.surface():
                    corrected_words.append((morpheme.surface(), normalized_form))
        return corrected_words

def apply_normalizing_function(df, normalizer)-> tuple:
    """google style docstring
    Args:
        df (pd.DataFrame): The DataFrame to normalize.
        normalizer (SudachiNounNormalizer): The normalizer object to use.
    Returns:
        tuple: A tuple containing the normalized DataFrame, a list of words to normalize, and a DataFrame of words to normalize."""
    # First check which words need to be normalized
    words_to_normalize = []
    for index, row in df.iterrows():
        # majorcategory = getattr(row, 'majorcategory') パス
        title = getattr(row, 'title')
        for column in df.columns:
            # if column == 'content' or column == 'tableinfo' or column == 'graphinfo':
            if column == 'content' or column == 'tableinfo' or column == 'graphinfo' or column == 'title':
                # Retrieve list of dicts
                cell_val = row[column]
                list_of_dicts = [ast.literal_eval(item) for item in cell_val]

                for dict in list_of_dicts:
                    # Check if 'file_name' key exists in the dictionary
                    if 'file_name' in dict:
                        # Check which words need to be normalized
                        extracted_corrected_words = normalizer.extract_corrected_words(dict['content'])
                        if extracted_corrected_words:
                            # words_to_normalize.append({dict['file_name'] : extracted_corrected_words, 'id' : dict['id']})
                            # words_to_normalize.append({'title': title, 'id': dict['id'], 'corrected_words': extracted_corrected_words})
                            string_extracted = str(extracted_corrected_words)
                            #words_to_normalize.append({'filename' : dict['file_name'], 'majorcategory' : majorcategory,  'corrected_words': string_extracted})
                            words_to_normalize.append({'filename' : dict['file_name'],  'corrected_words': string_extracted})
                        # Normalize words
                        dict['content'] = normalizer.normalize(dict['content'])
                converted_back = convert_dictlist_to_stringlist(list_of_dicts)
                df.at[index, column] = converted_back

    words_to_normalize_df = pd.DataFrame(words_to_normalize)
    return (df, words_to_normalize, words_to_normalize_df)


normalizer = SudachiNounNormalizer(dict_type="core", split_mode="C")
normalized_result = apply_normalizing_function(kebatori_df, normalizer)
sudachi_df = normalized_result[0]
words_to_normalize = normalized_result[1]
words_to_normalize_df = normalized_result[2]
display(sudachi_df.head())
```
